```python
import torch
import torchvision.datasets as dsets
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import random

s=0

for i in range(1000):
  print("Attempt:",i+1)
  training_epochs = 1
  batch_size = 100

  mnist_train = dsets.MNIST(root='MNIST_data/',train=True,transform=transforms.ToTensor(),download=True)#훈련 데이터

  mnist_test = dsets.MNIST(root='MNIST_data/',train=False,transform=transforms.ToTensor(),download=True)#테스트 데이터

  data_loader = torch.utils.data.DataLoader(dataset=mnist_train,batch_size=batch_size,shuffle=True,drop_last=True)
                                                                                      #=>데이터를 섞는다
  linear = torch.nn.Linear(784, 10, bias=True).to(device)

  criterion = torch.nn.CrossEntropyLoss().to(device)
  optimizer = torch.optim.SGD(linear.parameters(), lr=0.1)

  with torch.no_grad():
      X_test = mnist_test.test_data.view(-1, 28 * 28).float().to(device)
      Y_test = mnist_test.test_labels.to(device)

      prediction = linear(X_test)
      correct_prediction = torch.argmax(prediction, 1) == Y_test
      accuracy = correct_prediction.float().mean()
      print('Accuracy:', accuracy.item())

      r = random.randint(0, len(mnist_test) - 1)
      X_single_data = mnist_test.test_data[r:r + 1].view(-1, 28 * 28).float().to(device)
      Y_single_data = mnist_test.test_labels[r:r + 1].to(device)

      print('Label: ', Y_single_data.item())

      plt.imshow(mnist_test.test_data[r:r + 1].view(28, 28), cmap='Greys', interpolation='nearest')
      plt.show()

  for epoch in range(training_epochs):
      avg_cost = 0
      total_batch = len(data_loader)

      for X, Y in data_loader:
          X = X.view(-1, 28 * 28).to(device)
          Y = Y.to(device)

          optimizer.zero_grad()
          hypothesis = linear(X)
          cost = criterion(hypothesis, Y)
          cost.backward()
          optimizer.step()

          avg_cost += cost / total_batch

      print('Epoch:', '%04d' % (epoch + 1), 'cost =', '{:.9f}'.format(avg_cost))

  print('Learning finished')
  with torch.no_grad():
      X_test = mnist_test.test_data.view(-1, 28 * 28).float().to(device)
      Y_test = mnist_test.test_labels.to(device)

      prediction = linear(X_test)
      correct_prediction = torch.argmax(prediction, 1) == Y_test
      accuracy = correct_prediction.float().mean()

      X_single_data = mnist_test.test_data[r:r + 1].view(-1, 28 * 28).float().to(device)

      single_prediction = linear(X_single_data)
      print('Prediction: ', torch.argmax(single_prediction, 1).item())

  if torch.argmax(single_prediction,1)==Y_single_data.item():
    s+=1
  print("Aqquracy:",s/(i+1)*100,"%")

print("Right:",s)
print("Wrong:",1000-s)
print("Aqquracy:",s/i*100,"%")
```
